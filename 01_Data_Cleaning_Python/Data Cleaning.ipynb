```python
df_clean = df.copy()
```

                                          Data Cleaning

```python
df_clean[df_clean["SCHEDULED_TIME"].isna()]
```

```python
df_clean = df_clean.dropna(subset=["SCHEDULED_TIME"])
```

```python
df_clean["SCHEDULED_TIME"].isna().sum()
```

Rows with missing SCHEDULED_TIME represented only 6 records (<0.001% of the dataset) and corresponded to cancelled/diverted flights. These records were removed because the missing scheduled duration could not be reliably inferred and their removal had negligible impact on the analysis.

```python
delay_cols = [
    "AIR_SYSTEM_DELAY",
    "SECURITY_DELAY",
    "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY",
    "WEATHER_DELAY"
]

df_clean[delay_cols] = df_clean[delay_cols].fillna(0)

df_clean[delay_cols].isna().sum()
```

Missing values in delay reason columns were replaced with 0 because they represent absence of that specific delay type rather than unknown data.

                                     Feature Engineering

```python
df_clean["FLIGHT_STATUS"] = df_clean["CANCELLED"].map({0: "Completed", 1: "Cancelled"})
df_clean.loc[df_clean["DIVERTED"] == 1, "FLIGHT_STATUS"] = "Diverted"
```

```python
df_clean["FLIGHT_STATUS"].value_counts()
```

```python
df_clean["DEP_HOUR"] = df_clean["SCHEDULED_DEPARTURE"] // 100
df_clean["DEP_HOUR"].describe()
```

```python
def time_bucket(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df_clean["TIME_BUCKET"] = df_clean["DEP_HOUR"].apply(time_bucket)

df_clean["TIME_BUCKET"].value_counts()
```

```python
df_clean["ROUTE"] = (
    df_clean["ORIGIN_AIRPORT"].astype(str)
    + " → "
    + df_clean["DESTINATION_AIRPORT"].astype(str)
)
```

```python
df_clean["ROUTE"].head()
```

```python
df_clean["ROUTE"].nunique()
```

```python
def delay_category(x):
    if x <= 0:
        return "On Time"
    elif x <= 15:
        return "Minor Delay (1–15 min)"
    elif x <= 60:
        return "Moderate Delay (15–60 min)"
    else:
        return "Severe Delay (>60 min)"

df_clean["DELAY_CATEGORY"] = df_clean["ARRIVAL_DELAY"].apply(delay_category)

df_clean["DELAY_CATEGORY"].value_counts()
```

```python
#Q1: Which airline has highest average delay?
airline_delay = df_clean.groupby("AIRLINE")["ARRIVAL_DELAY"].mean().sort_values()
airline_delay
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

airline_delay.sort_values().plot(
    kind="barh",
    figsize=(10,6),
    color="steelblue"
)

plt.title("Average Arrival Delay by Airline (2015)")
plt.xlabel("Delay (minutes)")
plt.ylabel("Airline")

plt.tight_layout()
plt.show()
```

```python
#What causes flight delays?
delay_cols = [
    "AIR_SYSTEM_DELAY",
    "SECURITY_DELAY",
    "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY",
    "WEATHER_DELAY"
]

delay_summary = df_clean[delay_cols].sum().sort_values(ascending=False)
delay_summary
```

```python
delay_summary = df_clean[[
    "AIR_SYSTEM_DELAY",
    "SECURITY_DELAY",
    "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY",
    "WEATHER_DELAY"
]].sum()

plt.figure(figsize=(7,7))
plt.pie(
    delay_summary,
    labels=delay_summary.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Delay Causes Distribution (Total Delay Contribution)")
plt.axis("equal")  # makes it a perfect circle
plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    delay_category_dist,
    labels=delay_category_dist.index,
    autopct="%1.1f%%",
    startangle=140,
    wedgeprops={"edgecolor": "black"}
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Flight Delay Severity Distribution")
plt.axis("equal")
plt.show()
```
